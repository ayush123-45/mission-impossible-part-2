## 📊 *The EDA Framework*
> “Data speaks, but only if you know how to listen.”

The Exploratory Data Analysis phase of this project is designed to bridge the gap between raw text and actionable machine learning features. Before a single model is trained, we perform a deep-tissue scan of the IMDB corpus to identify the linguistic patterns that define human sentiment. Our EDA is governed by three pillars: *Distribution, **Composition, and **Correlation*.

### 🔍 *EDA Objectives & Methodology*
* ⚖️ *Statistical Distribution:* Verifying the 1:1 class balance and analyzing *Review Length Variance* to see if length correlates with sentiment.
* 🔤 *Textual Composition:* Performing *N-Gram Analysis* (Unigrams/Bigrams) and assessing *Stop-Word Impact* to reveal the core vocabulary of cinema critique.
* ☁️ *Sentiment Correlation:* Utilizing *Word Clouds* and measuring *Lexical Diversity* to identify if "Elite" reviewers use different language patterns than casual viewers.

---

## 🎯 *Technical Analytical Goals*

### *1. Feature Engineering*
Transforming raw text into mathematical vectors that machines can process:
* *Representations:* Comparing *Bag-of-Words, **TF-IDF, and **Word Embeddings*.
* *Visualization:* Using *t-SNE* or *PCA* to see how sentiment clusters in high-dimensional space.

### *2. Algorithmic Performance*
We pit classical statistical methods against modern deep learning:
* *Baselines & Ensembles:* Logistic Regression, Naive Bayes, and Random Forest.
* *Deep Learning:* LSTMs, CNNs, and *Transformer-based models*.

### *3. Evaluation & Insights*
Moving beyond "Accuracy" to understand why models succeed or fail:
* *Metrics:* Analysis via Precision, Recall, F1-Score, and ROC-AUC.
* *Challenges:* Discussing the impact of *sarcasm, **ambiguity, and **context*.

```python
#--- Optimized Cleaning + Diagnostics for TMDB ---

# Cleaning (same as before, but lighter)
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['runtime']      = pd.to_numeric(df['runtime'], downcast='integer')
df['budget']       = pd.to_numeric(df['budget'], downcast='integer')
df['revenue']      = pd.to_numeric(df['revenue'], downcast='integer')

# Missing values summary
missing     = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df  = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df  = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing %', ascending=False)

print("Missing Values Summary (Top 15):")
print(missing_df.head(15))

# Basic statistics (numeric + categorical separately)
print("\nNumeric Statistics:")
print(df.describe().T.round(2))

print("\nCategorical Statistics (Top 20):")
print(df.describe(include='object').T.head(20))

# Visualization (Top 15 missing only)
plt.figure(figsize=(12,6))
sns.barplot(x=missing_df.head(15).index, y=missing_df.head(15)['Missing %'], palette="viridis")
plt.xticks(rotation=90)
plt.title('Top 15 Columns with Missing Values')
plt.ylabel('Missing %')
plt.xlabel('Columns')
plt.show()
```

```python
# Adjust dtypes for TMDB Movies Daily Updates Dataset

# Convert release_date to datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Cast numeric counts to integers
df['vote_count']   = df['vote_count'].astype('Int64')
df['imdb_votes']   = df['imdb_votes'].astype('Int64')
df['runtime']      = df['runtime'].astype('Int64')

# Cast financials to integers
df['budget']       = df['budget'].astype('Int64')
df['revenue']      = df['revenue'].astype('Int64')

# Ratings remain float (decimal precision is useful)
df['vote_average'] = df['vote_average'].astype('float64')
df['imdb_rating']  = df['imdb_rating'].astype('float64')
df['popularity']   = df['popularity'].astype('float64')

# Identifiers
df['id']           = df['id'].astype('Int64')
df['imdb_id']      = df['imdb_id'].astype('string')

# Textual / categorical fields remain string/object
text_columns = [
    'title', 'status', 'original_language', 'original_title', 'overview',
    'tagline', 'genres', 'production_companies', 'production_countries',
    'spoken_languages', 'cast', 'director', 'director_of_photography',
    'writers', 'producers', 'music_composer', 'poster_path'
]
df[text_columns] = df[text_columns].astype('string')

print("✅ Dtypes adjusted successfully!")
print(df.dtypes)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Numeric Pipeline: Fill missing values with median and scale
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Categorical Pipeline: Fill missing with 'most_frequent' and encode
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

# Combine into a single preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'regressor__n_estimators': [100, 200],
    'regressor__max_depth': [None, 10, 20]
}

grid = GridSearchCV(
    Pipeline([('preprocessor', preprocessor),
              ('regressor', RandomForestRegressor(random_state=42))]),
    param_grid,
    cv=5,
    scoring='r2'
)

grid.fit(X_train, y_train)
print("Best params:", grid.best_params_)
print("Best R²:", grid.best_score_)
```
